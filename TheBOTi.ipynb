Este primer script pretende leer el histórico de $BTC y encontrar divergencias

Un valor clave que vamos a calcular es RSI; 
```R

RSI = 100 - 100/(1+RS)
```
Donde RS es la ganancia media en un periodo de subidas/ perdida media de periodos de bajada en un marco de tiempo determinado

Los datos son el historico hasta el 6 de abril sacados de http://api.bitcoincharts.com/v1/csv/

bajado el bitstamp

````r
#install.packages("anytime")
#install.packages("TTR") #https://cran.r-project.org/web/packages/TTR/TTR.pdf
library(anytime)
library(TTR)
````

````r
cabecero =c( "unixtime","price","amount")
````

````r
dataframe =read.csv("data/oddbtcdata2") 
````

````r
colnames (dataframe) <- cabecero
````

````r
head(dataframe)
````

````r
plot (anytime(dataframe$unixtime), dataframe$price,  type="l")
````

````r
rsi <- RSI(dataframe$price)
````

````r
plot(rsi) #Modo basico, ver documentacion para usarlo con criterio
````
